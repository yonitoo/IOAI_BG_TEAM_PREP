## Операции

Може да мислите за елементите като матрици, но трябва да знаете, че операциите с тях се **прилагат  поелементно (elementwise)**:

```python
i = np.eye(4)
i
```

```python
random_numbers = np.random.rand(4, 4)
random_numbers
```

Произведението на тези две матрици трябва да даде `random_numbers` (`i` е единична матрица). Резултата, обаче, е друг:

```python
np.eye(4) * np.random.rand(4, 4)
```

Събирането е същото:

```python
i + random_numbers
```

Както и повдигането на степен:

```python
random_numbers ** i
```

Ако единия операнд е скалар, операцията се "векторизира" – изпълнява се за всеки елемент:

```python
random_numbers + 1
```

```python
(random_numbers * 10).astype(np.int64)
```

```python
random_numbers - 0.5
```

Резултата може и да е друг тип:

```python
random_numbers >= 0.5
```

Като друго забавление, нека видим интеграцията:

```python
from matplotlib.patches import Polygon

def draw_sine_integration_example():
    func = np.sin
    a, b = np.pi / 2, np.pi
    x = np.linspace(-0.2, np.pi * 1.5)
    y = func(x)

    fig, ax = plt.subplots(figsize=(10, 6))
    plt.plot(x, y, 'r', linewidth=2)
    plt.grid()

    ix = np.linspace(a, b)
    iy = func(ix)
    verts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)]
    poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
    ax.add_patch(poly)

    ax.axhline(color='k')
    ax.axvline(color='k')
    ax.annotate(r'$\frac{\pi}{2}$', (np.pi / 2, 0), xytext=(0, -20), textcoords='offset points', size='x-large')
    ax.annotate(r'$\pi$', (np.pi, 0), xytext=(0, -20), textcoords='offset points', size='x-large')
```

```python
draw_sine_integration_example()
```

$$
\begin{align*}
\int_{\pi / 2}^{\pi}\sin(x)\mathrm{d} x &= - \cos(x) \Big|_{\pi / 2}^\pi \\
&= - \cos(\pi) - \big(- \cos(\frac{\pi}{2}) \big) \\
&= \cos(\frac{\pi}{2}) - \cos(\pi) \\
&= 0 - (-1) \\
&= 1
\end{align*} 
$$

## Съединения (Joins)

Има и операции, подобни на съединенията (JOIN) в базите данни:

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})

df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
```

```python
df1
```

```python
df2
```

 [`pandas.merge()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.merge.html) прави стандартен join (inner join в SQL):

```python
pd.merge(df1, df2)
```

По подразбиране се ползват съвпадащите колони. Може да го направим експлицитно:

```python
pd.merge(df1, df2, on=['key'])
```

По подразбиране се прави inner join – ако някой ключ го няма в един от двата `DataFrame`-а, няма да го има и в резултата. Ако искаме пълно съединение (`outer join`), трябва да го кажем експлицинто. Тогава `merge` ще азапази реда, попълвайки стойностите с `NaN` (т.е. празни стойности).

Аналогично могат да се правят и left join, right join, cross join. 

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

Съществува и функция [`DataFrame.join()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html), която прави подобни неща.

```python
func = np.sin
a, b = np.pi / 2, np.pi
x = np.linspace(-0.2, np.pi * 1.5 + 0.2)
y = func(x)

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'r', linewidth=2)
plt.grid()

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.xaxis.set_ticks(np.arange(0, 5, np.pi / 4))

ix = np.linspace(a, b)
iy = func(ix)
verts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

ax.annotate(r'$\frac{\pi}{2}$', (np.pi / 2, 0), xytext=(0, -20), textcoords='offset points', size='x-large')
ax.annotate(r'$\pi$', (np.pi, 0), xytext=(0, -20), textcoords='offset points', size='x-large');
```
