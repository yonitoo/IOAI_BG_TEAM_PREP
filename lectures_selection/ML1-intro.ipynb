# Машинно самообучение 

Машинното самообучение е предмента област, подобласт на изкуствения интелект (AI). В нея се включват всички алгоритми, които се обучават да решават различни задачи, чрез съставяне на статистически модели над множество от данни. 

В повечето случаи, обучението на повече данни води до подобряване на точността на алгоритъма.

Възможни са приложения като: автоматизирани диагностични процедури, откриване на измами с кредитни карти, анализ на фондовия пазар,  разпознаване на глас, визуални образи и текст.

Проблемите, подходящи за машинно самообучение, имат следните качества:

* Работи се най-често с големи масиви от данни, но има и изключения, с които трябва да се справяме.
* "Нагаждаме" различни алгоритми към проблема, докато постигнем добър резултат.
* Тренираме модел, който ни позволява да отговорим на въпроси за нови данни.
* Rule of thumb: ако човек може да реши проблема за около 2 секунди, вероятно може да направим ML решение.

Някои интересни наблюдения:

* Повече данни могат да ни помогнат да постигнем по-добър резлутат (до един момент).
* При коренно различни данни може да се наложи да ползваме друг алгоритъм за същия проблем.
* Процесът е до голяма степен екпериментален - на принципа проба-грешка. 

# Математика

Каква математика ще ви е нужна? Кратката версия:

$$ y = ax + b $$

(където $y$ и $b$ са вектори, а $a$ и $x$ са матрици).

* y: това, което искаме да прогнозираме (например цена на кола).
* X: данни, които използваме за прогноза (цвят, двигател, салон).
* а (наклон): как зависи цената от данните Х (например с колко се покачва цената на колата, ако двигателят е с една единица по-мощен).
* b (пресечна точка): каква ще е цената, ако двигателя го няма.

Малко по-детайлно:

Задължително ще ви трябва да разбирате от поне малко линейна алгебра. Като начало, трябва да разбирате от умножение на матрици и да ви е относително комфортно да го правите. Това е едно добро начало.

В подробности – всеки алгоритъм си има особеностите и математиката, свързана с него. Линейната регресия е напълно разбираема с не много дълбоко разбиране на математика. Други алгоритми като Support Vector Machines или Principal Component Analysis са по-сложни и искат повече познания. На практика, може да стигнете доста далеч с повърхностно разбиране на тези алгоритми. На теория, колкото повече математика знаете, толкова по-добре ще се оправите.

Ние ще гледаме на нещата от по-практическа гледна точка като изградим добра интуиция за различните алгоритми как работят.

# Supervised vs. unsupervised learning

Алгоритмите могат да се разделят на два видя:

* Supervised learning – такива, които разполагат с етикирани данни и генерализират (да отговарят на въпроси за нови данни).
* Unsupervised learning – такива, които нямат етикети и трябва да открият статистически зависимости в данните.

# Supervised learning ("обучение с учител")

Примери за supervised learning са:

* Регресия - предвиждане на непрекъснати числови стойности: При набор от данни с цени и параметри на апартаменти да определим колко би струвал друг апартамент с определени апартаменти.
* Класификация - предвиждане на булеви или цели числени стойности: При набор от данни за тумори да определим дали един е доброкачествен или злокачествен.

# Unsupervised learning ("обучение без учител")

Тези алгоритми са по-разнородни и приложими в определени сфери. Например:

* Клъстеризация: При набор от потребители и техните филмови рейтинги да създадем групи от видове предпочитания.
* Намаляне на размерността: При набор от многомерни данни данни да сведем броя измерения до по-малък такъв запазвайки повечето информация.

# Регресия vs. Класификация

Бихме могли да разделим supervised learning на два вида:

1. Регресия – опитваме се прогнозираме непрекъсната стойност. Искаме да намерим права, която най-добре описва данните.
* Пример: Цена на апартамент, температура утре.

2. Класификация – опитваме се да определим дали данните попадат в една от няколко дискретни категории (класове). Ако са 2 категориите, наричаме я бинарна класификация. (напр. доброкачествен или злокачествен тумор)

Как работи класификацията с 3+ класа? Често се използва подход, наречен one-vs-many (един срещу всички). За всеки клас моделът се учи да отговаря на въпроса: "Този обект принадлежи ли към този конкретен клас или към някой от останалите?". Накрая избираме класа, за който моделът е най-сигурен.

# К най-близки съседи (k-Nearest Neighbours, kNN)

### Supervised, класификация

Това е може би най-интуитивният алгоритъм.

Как работи? Той не "учи" по традиционния начин, а просто запаметява целия набор от данни. Когато получим нова точка, за да я класифицира, алгоритъмът:

1. Намира К-те най-близки до нея точки от данните за учене (измерено с евклидово разстояние).
2. Проверява към кои класове принадлежат тези "съседи".
3. "Гласува" и присвоява на новата точка класа, който е най-често срещан сред съседите.

## Интуиция:

Новодошли сме в квартала и искаме да разберете дали съседите ни са фенове на Левски или ЦСКА. Питаме най-близките си 3-ма съседи (k=3). Ако двама са за Левски и един за ЦСКА, залагаме, че в тази част на квартала преобладават феновете на Левски.

Не правете това вкъщи. :D

Броят съседи определя колко "гладка" е границата между двата класа. Ето един пример с различен брой съседи:

````python
from sklearn.neighbors import KNeighborsClassifier

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} neighbour(s)".format(n_neighbors))
````

Фонът в горната диаграма показва в кой от двата класа ще попадне всяка нова точка.

* При k=1, границата е много начупена и следва данните перфектно. Моделът е твърде сложен и рискува да "наизусти" данните, включително и случайния шум (overfitting).
* При k=9, границата е много по-плавна. Моделът е по-прост и по-добре обобщава (генерализира) общата тенденция, макар и да класифицира някои точки грешно. Това често е предимство, защото игнорира аномалии в данните.

## Overfitting vs underfitting

* Overfitting - когато моделът научи тренировъчните данни твърде добре, вкл. шум и outliers. Тогава не се справя добре с прогноза на нови данни.
* Underfitting - когато моделът е твърде прост да прихване тертипите/pattern-ите в данните - води до лошо представяне и на тренировъчните данни, и на нови такива.

## Интуиция

Представете си, че от всяка точка пускате вертикална отсечка до правата. Алгоритъмът се опитва да намести правата така, че общата "площ" на всички квадрати, построени върху тези отсечки, да е възможно най-малка.

В горния случай е трудно да се направи по-точен линеен модел, тъй като данните имат голяма вариация за едни и същи входни стойности. При наличието на повече измерения обикновено може да се постигне по-добър резултат.

# Principal Component Analysis (PCA)
### Unsupervised, намаляване на размерността (dimensionality reduction) 

Това е техника, която ни позволява да сведем многомерно пространство до такова с по-малко измерения, като се стремим да запазим възможно най-много информация (вариация).

## Интуиция:

Представете си, че осветявате облак от точки в 3D пространството с фенерче. PCA намира ъгъла, под който да осветите точките, така че тяхната сянка (проекция) върху стената да е възможно най-разпръсната и информативна. Тази сянка е представяне на данните в по-ниска размерност.

````python
mglearn.plots.plot_pca_illustration()
````

Графиките горе илюстрират как PCA се опитва да сведе двуизмерно пространство до едноизмерно. Данните могат да се представят чрез базис от два вектора (component 1 и component 2), където component 1 съдържа много информация, а component 2 – малко. С този dataset бихме могли да тренираме относително точен модел само с component 1. Обърнете внимание, че той е функция на оригиналните два feature-а.

При две измерения това не е нужно, но при 1000+ подобна трансформация на данните може да е задължителна за да има шанс да съберете модела в паметта.

## Друг метод за намаляване на размерността: Autoencoders

Autoencoders са вид невронни мрежи, които се използват за същата цел, но по по-мощен, нелинеен начин. Те се състоят от две части: Encoder, който компресира данните до ниска размерност (т.нар. Принуждавайки мрежата да възстанови данните от компресираното представяне, ние я караме да се научи кои са най-важните, съществени признаци в тях.


# k-Means Clustering
### Unsupervised, клъстеризация

Това е най-популярният алгоритъм за клъстеризация - задача, при която се опитваме да разделим данните на определен брой групи (клъстери) без да имаме предварителни етикети.

### Алгоритъмът работи по следния начин итеративно

1. Инициализация: Избира K на брой случайни точки, които наричаме центроиди (центрове на клъстери).
2. Присвояване: Всяка точка от данните се присвоява (оцветява) към най-близкия до нея центроид.
3. Актуализация: Позицията на всеки центроид се преизчислява като средно аритметично на всички точки, които са присвоени към него.
4. Повтаряне: Стъпки 2 и 3 се повтарят, докато центроидите спрат да се местят значително.

````python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# генерираме данни
# X, y = make_blobs()
X, y = make_blobs(random_state=1)

# 3 клъстера
kmeans = KMeans(n_clusters=3, n_init=1)
kmeans.fit(X)

# визуализираме
plt.figure(figsize=(10, 6))
mglearn.discrete_scatter(X[:, 0], X[:, 1], kmeans.labels_, markers='o')
mglearn.discrete_scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], [0, 1, 2], markers='^', markeredgewidth=2)
plt.title("Резултат от k-Means клъстеризация")
plt.show()
````

````python
mglearn.plots.plot_kmeans_algorithm()
````

Този алгоритъм е недетерминистичен – различен избор на първоначални точки може да произведе различни резултати. По тази причина на практика се изпълнява няколко пъти и се взема добър резултат.

Това също е мотив, който се среща често.

### Други алгоритми за клъстеризация

k-Means е чудесен, но предполага, че клъстерите са сферични и с еднаква плътност. Други алгоритми се справят с различни структури:

- DBSCAN: Намира клъстери на базата на плътност. Може да открива клъстери с произволна форма и автоматично да идентифицира точки, които не принадлежат на никой клъстер (шум). Не изисква предварително задаване на броя клъстери.
- Agglomerative Clustering (агломеративно клъстеризиране): Йерархичен метод, който започва с всяка точка в собствена групичка и итеративно обединява най-близките клъстери, докато не остане само един.

````python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# генерираме си точки
X, y = make_moons(n_samples=250, noise=0.1, random_state=0)

# създаваме и обучаваме DBSCAN модела
#    - eps=0.1: Ако две точки са на по-малко от 0.1 разстояние, са съседи.
#    - min_samples=5: Нужни са поне 5 точки, за да се формира клъстър.
dbscan = DBSCAN(eps=0.1, min_samples=5)
# dbscan = DBSCAN(eps=0.3, min_samples=5) # TODO: Какво ще се промени тук?
clusters = dbscan.fit_predict(X)

# визуализираме
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", s=50)
plt.title("DBSCAN клъстеризация на Moons")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````

````python
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_moons

# генерираме си данни
X, y = make_moons(n_samples=250, noise=0.1, random_state=0)

# създаваме и обучаваме модела
#    - n_clusters=2: Казваме му да спре, когато останат 2 клъстера.
agg_cluster = AgglomerativeClustering(n_clusters=6)
clusters = agg_cluster.fit_predict(X)

# визуализираме
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap="viridis", s=50)
plt.title("Агломеративно клъстеризиране на Moons")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````
